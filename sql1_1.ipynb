```python
# MySQL 접속

import pymysql
```

```python
conn = pymysql.connect(
    host="localhost",
    user="root",
    password="class1234",
    database="test",
    charset="utf8"
)

cursor = conn.cursor() # 연결된 DB에 SQL을 실행하는 객체

print("MySQL 연결 성공")
```

```python
# INSERT

products = [
    {"name": "키보드", "price": 50000},
    {"name": "마우스", "price": 25000},
    {"name": "모니터", "price": 300000},
    {"name": "노트북", "price": 1200000},
    {"name": "USB", "price": 15000},
    {"name": "웹캠", "price": 80000},
    {"name": "헤드셋", "price": 90000},
    {"name": "스피커", "price": 70000},
    {"name": "외장하드", "price": 140000},
    {"name": "프린터", "price": 220000}
]
```

```python
sql = """
insert into product(name, price)
values(%s, %s)
"""
```

```python
for i, p in enumerate(products):
    cursor.execute(sql, (p["name"], p["price"]))
    # execute(str, tuple) 채울 표현식%가 없으면 두번째 인수는 생략, 하나라도 있으면 항상 튜플로 순서대로 입력
    print(f"{i+1}번째 insert 실행")

conn.commit()
```

```python
# SELECT

sql = """
select * from product
"""

cursor.execute(sql)

rows = cursor.fetchall()
print(type(rows)) # <class 'tuple'>

for r in rows:
    print(r)
```

```python
# UPDATE
# id = 1인 모니터의 price null → 150000원으로 수정

update_price_val = 150000
update_id = 1


sql = """
update product
set price = %s
where id = %s
"""

cursor.execute(sql, (update_price_val, update_id))

conn.commit()

print(f"id {update_id}번 수정 완료")
```

```python
# DELETE
try:
    sql = """
    delete from product
    where price is null
    """

    # db null값을 연산할때는 db에서 제공하는 null 연산자만을 사용해야 한다.
    # is null, is not null, ... : 사용 가능
    # =, != : 사용 불가

    cursor.execute(sql)
    conn.commit()

    # select 는 실행 후 cursor에 반환된 결과셋 가진다.
    # insert/ update / delete 는 cursor에 반환된 실행 결과 행의 수를 가진다.
    print(f"{cursor.rowcount}행 삭제 완료")

except:
    print("db 예외 발생")
    
finally:
    cursor.close()
    conn.close()
```

```python
# MySQL DB연결종료

cursor.close()
conn.close()

print("연결 종료")
```
